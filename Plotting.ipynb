```python
#impot API
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("co2_emission.csv") #importing data
data_us = data[19710:19928]
data_chn = data[4280:4485]
```

```python
fig, ax = plt.subplots() #subplot outputs two variables: figure and axis
ax.plot(data_us["Year"].values, data_us["CO2"].values) #arguments: first x data, second y data, marker, color, linestyle
ax.set_xlabel("Year") #labels x-axis
ax.set_ylabel("CO2 Emissions in Trillions") #labels y-axis
ax.set_title("Yearly CO2 Emissions in US") #titles the plot
ax.set_xticklabels(data_us["Year"], rotation = 90) #rotates the x-axis labels by 90 degrees
plt.show() #draw the plot
```

```python
#Unclogging plots by splitting up into multiple plots
fig, ax = plt.subplots(2, 1, sharey = True) #2 rows by 1 column of plots, share y makes so y increments are same
ax[0].plot(data_us["Year"], data_us["CO2"], color = 'b') #set first 
ax[1].plot(data_chn["Year"], data_chn["CO2"], color = 'r')
plt.show()
```

```python
#twin axes
fig, ax = plt.subplots()
ax.plot(data_us["Year"], data_us["CO2"], color = "blue", label = "US") #plot US data as blue line, label in legend
ax.set_xlabel("Year")
ax.set_ylabel("Yearly Carbon Emissions in the US", color = "blue") #make US y label blue
ax.tick_params(colors = "blue") #make ticks and numbers blue
ax2 = ax.twinx() #creates a twin axis
ax2.plot(data_chn["Year"], data_chn["CO2"], color = "red", label = "China") #plot China data as red line, label legend
ax2.set_ylabel("Yearly Carbon Emissions in China", color = "red") #make China y label red
ax2.tick_params(colors = "red") #make ticks and numbers
fig.legend()
ax2.plot()
```

```python
#using a function to reduce repitition in code
def plot(axes, x, y, color, xlabel, ylabel, title):
    axes.plot(x, y, color = color) #plotting the inputs of x, y
    axes.set_xlabel(xlabel) #setting the x-axis label
    axes.set_ylabel(ylabel, color = color) #setting the y-axis label
    axes.tick_params('y', colors = color) #setting the color of the ticks and nubmers
    axes.set_title(title) # set title
fig, ax = plt.subplots()
plot(ax, data_us["Year"], data_us["CO2"], "blue", "Year", "Yearly Carbon Emissions in the US", "US")
```

```python
#annotation https://matplotlib.org/stable/tutorials/text/annotations.html
ax2.annotate(">1 degree", xy = (pd.Timestamp("2008-10-06"), 1), )
```

```python
#bar charts
medals = pd.read_csv('medals_by_country_2016.csv', index_col = 0) #read
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"]) #plot a bar chart
ax.bar(medals.index, medals["Silver"], bottom = medals["Gold"]) #Stacked bar chart, bottom of silver is gold
ax.bar(medals.index, medals["Bronze"], bottom = medals["Gold"] + medals["Silver"]) #Stacked, bottom silver + gold
plt.show
```
