```python
data2 = pd.read_csv('monthlyCSV.csv', header = 0, index_col = 0)
data2 = data2.drop(range(2007, 2021))
npdata2 = data2.to_numpy()
g2 = []
for i in npdata2:
    for j in i:
        if not np.isnan(j):
            g2.append(j)
min_y = min(g2)
max_y = max(g2)
for i in range(len(g2)):
    g2[i] -= min_y
    g2[i] /= max_y - min_y
```

```python
prev_50 = []
prev_25 = []
prev_10 = []
cur = g2[-600:-300]
av_50 = sum(cur)/300
av_25 = sum(cur[-300:-120])/180
av_10 = sum(cur[-120:])/120
for i in range(len(g2)-600):
    prev_10.append(av_10)
    prev_25.append(av_25)
    prev_50.append(av_50)
    cur.append(g2[600+i])
    av_50 = sum(cur[-600:-300])/300
    av_25 = sum(cur[-300:-120])/180
    av_10 = sum(cur[-120:])/120
prev_10.append(av_10)
prev_25.append(av_25)
prev_50.append(av_50)
print(len(prev_50))
```

```python
min_y = min(g)
max_y = max(g)
for i in range(len(g)):
    g[i] -= min_y
    g[i] /= max_y - min_y
g
```

```python
def create_inout_sequences(input_data, tw = 36):
    global prev_10, prev_25, prev_50
    inout_seq = []
    L = len(input_data)
    for i in range((L-tw-tw)//12):
        train_seq = input_data[i*12:(i*12)+tw]
        train_seq.extend([prev_10[i*12], prev_25[i*12], prev_50[i*12]])
        train_label = input_data[(i*12)+tw:(i*12)+tw+tw]
        inout_seq.append((train_seq ,train_label))
    return inout_seq

train_inout_seq = create_inout_sequences(g)
```

```python
import torch.nn as nn
from torch.nn import Sequential
import torch as tr
tr.manual_seed(0)
```

```python

model = Sequential( #BP, backpropagation
    
    nn.Linear(63, 128),
    nn.BatchNorm1d(1),
    nn.ReLU(),
    
    nn.Linear(128, 64),
    nn.BatchNorm1d(1),
    nn.ReLU(),

    nn.Linear(64, 60),
    nn.Sigmoid(),
)
```

```python
class RNN(nn.Module):
    def __init__(self, input_size):
        super(RNN, self).__init__()
        self.rnn = nn.LSTM(
            input_size = input_size,
            hidden_size = 128, #number of features coming out of the LSTM layer
            num_layers = 2, #number of LSTM layers
            batch_first = True,
            
        )
        self.out = nn.Sequential(
            nn.Linear(128, 64),
            nn.BatchNorm1d(1),
            nn.ReLU(),
            
            nn.Linear(64, 32),
            nn.BatchNorm1d(1),
            nn.ReLU(),
            
            nn.Linear(32,36),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        r_out, (h_n, h_c) = self.rnn(x, None) #initial h and c is 0
        out = self.out(r_out)
        return out
```

```python
model = RNN(39)
loss_function = nn.MSELoss()
op = tr.optim.Adam(model.parameters(), lr=0.0001)
#op_2 = tr.optim.Adam(model.parameters(), lr = 0.0001)

print(model)
```

```python
device = tr.device("cuda" if tr.cuda.is_available() else "cpu")
model = model.to(device) #looks for gpu if gpu is available and puts model into gpu for computation
repeats = 600 #decides how many times each piece of data is ran through the model
train_error = []
test_error = []
t_error = 0
for repeat in range(repeats):
    model.train() #mode change into training
    #if repeats > 200:
        #op = op_2
    for seq, labels in train_inout_seq[1:]: #enumerate returns data into data and index into index_train, for loop is a single test of all info
        #train_dataLoader includes 2 parts from the return, one is the target, other is data
        seq = [seq]
        labels = [labels]
        x = tr.Tensor(seq) #input
        y = tr.Tensor(labels) #label or target
        op.zero_grad() #clear previous derivative, otherwise would cause error
        pred = model(tr.unsqueeze(x, dim = 1)) #give the information (x) into the model, pred is the prediction of the model
        
        error = loss_function(pred,tr.unsqueeze(y, 1)) # Compares prediction with actual target with cross entropy
        error.backward() #pass the loss back (back propagation)
        op.step() #with the loss, can optimize the model
        train_error.append(error)
        
    seq, labels = train_inout_seq[1]
    seq = [seq]
    labels = [labels]
    x = tr.Tensor(seq)
    y = tr.Tensor(labels)
    op.zero_grad()
    pred = model(tr.unsqueeze(x, dim = 1))
    t_error = loss_function(pred, tr.unsqueeze(y, 1))
    test_error.append(t_error)
        
    print ("Epoch %d Loss: %f Test Loss: %f"%(repeat, error, t_error))



#% is a replacement, letter after decides format
```

```python

plt.plot(np.arange(len(test_error)), test_error)
plt.title('Test Error with Data from 2005 to Present')
plt.xlabel('Epoch Number')
plt.ylabel('Tested Error')
plt.savefig('drought_error.pdf')
```

```python
z = np.linspace(-6, 6, 100) 
def sigmoid(z):
    return 1/(1 + np.exp(-z))
def tanh(z):
    return 2/(1 + np.exp(-2*z)) - 1
def relu(z):
    a = []
    for i in z:
        if i < 0:
            a.append(0)
        else:
            a.append(i)
    return a

fig, ax = plt.subplots(1, 3)

a = sigmoid(z)
b = tanh(z)
c = relu(z)
ax[0].plot(z, a)
ax[0].set_ylim([-0.2,1.2])
ax[0].set_title("Sigmoid", size = 15)
ax[1].plot(z, b)
ax[1].set_ylim([-1.5, 1.5])
ax[1].set_title("Hyperbolic Tangent", size = 15)
ax[2].plot(z, c)
ax[2].set_ylim([-2, 10])
ax[2].set_title("ReLU", size = 15)
plt.savefig('activation.pdf')
```
